```
df = spark.read.format("csv").option("header","true").load("/mnt/Gen2/")
```

```
display(df)
```

```
sortDF = df.sort('C_CUSTKEY')
display(sortDF)
```

```
sortDF.write.mode("overwrite").format("parquet").option("path", "/tmp/customer/sortedData/").saveAsTable("sortedData")
df.write.mode("overwrite").format("parquet").option("path", "/tmp/customer/unsortedData").saveAsTable("unsortedData")
df.write.mode("overwrite").format("json").option("path", "/tmp/customer/jsonData").saveAsTable("jsonData")
```
